```python
# Making the imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
N=20
X=np.random.random(N)*2*np.pi
Y=np.sin(X)
```

```python
plt.scatter(X, Y)
plt.show()
```

```python
# Building the model
m = 0
c = 0
b = 0
L = 0.0001  # The learning Rate
epochs = 2000  # The number of iterations to perform gradient descent

n = float(len(X)) # Number of elements in X

# Performing Gradient Descent 
for i in range(epochs): 
    Y_pred = b*(X**2) + m*X + c  # The current predicted value of Y
    D_b = (-2/n) * sum((X**2) * (Y - Y_pred))
    D_m = (-2/n) * sum(X * (Y - Y_pred))  # Derivative wrt m
    D_c = (-2/n) * sum(Y - Y_pred)  # Derivative wrt c
    m = m - L * D_m  # Update m
    c = c - L * D_c  # Update c
    b = b - L * D_b  # Update b
    
print (m, c, b)
```

```python
# Making predictions
Y_pred = b*(X**2) + m*X + c

plt.scatter(X, Y) 
plt.scatter(X,Y_pred,color='g')
plt.show()
```

```python
mse=(1/N)*sum((Y-Y_pred)**2)
print(mse)
```
