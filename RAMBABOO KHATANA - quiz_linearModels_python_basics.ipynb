### Sum of squares
Given  a list of floats, find  the sum of squares <br>
Eg., 
l = [1, 2, 3] <br>
sum_sqs(l) == 14

```python
def sum_sqs(l):
  """
  Input:
    l : list of ints
    y : int, sum of squares of elements of l
  """
  # YOUR CODE HERE
  sum =0
  for i in l:
    sum = sum +i**2
  y=sum
  return y
```

```python
l = [1, 2, 3] 
sum_sqs(l) == 14
```
